## Dilation

With A and B as sets in $Z^2$, the *dilation* of A by B, denoted as $A \oplus B$, is defined as:

$$A \oplus B = \{ z | (\hat{B})_z \cap A \neq \varnothing \} $$

This can be understood to based on the reflection of B, shifted by z units. The dilation of A by B is the set of all *displacements* z, such that the reflection of B and A overlap by at least one element.

In the case of images, B is the *structuring element* and A is the image.

`scikit-image` has has a package of pre-made structuring elements that we can apply to our images. Let's take a look at a few:

```python
from skimage.morphology import square, rectangle, disk, star
from skimage.morphology import binary_dilation as dilate
from skimage.morphology import binary_erosion as erode
from skimage import data
import matplotlib.pyplot as plt
import numpy as np
```

```python
square(10)
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(12, 10))

for i, size in enumerate([4, 10, 20, 40]):
    ax[i].imshow(square(size), cmap='gray')
    ax[i].set_xlim([-10, 50])
    ax[i].set_ylim([-10, 50])
```

To demonstrate, what happens is we try to dilate the largest structuring element by the smallest?

```python
back = np.zeros((100,100))
back[10:50, 10:50] = square(40)
back = back

test = dilate(back, selem=square(10))

fig, ax = plt.subplots(ncols=2, figsize=(6,6))
ax[0].imshow(back, cmap='gray')
ax[0].set_xlim([0,60])
ax[0].set_ylim([0,60])

ax[1].imshow(test, cmap='gray')
ax[1].set_xlim([0,60])
ax[1].set_ylim([0,60])
```

Dilations are usually performed to bridge gaps in images. Let's try an example image, to see how it works.

```python
image = data.camera()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

axes[0, 0].imshow(image, cmap='gray')
binarim = image > 50
axes[0, 1].imshow(binarim, cmap='gray')
axes[1, 0].imshow(1-dilate(1-binarim, selem=square(4)), cmap='gray')
axes[1, 1].imshow(1-dilate(1-binarim, selem=square(10)), cmap='gray')
for ax in axes:
    for a in ax:
        a.axis('off')
```

Note that the square structing element is symmetric in x and y, but it doesn't have to be. Try a rectangle to have different rates of dilation in x and y:

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

axes[0, 0].imshow(image, cmap='gray')
binarim = image > 50
axes[0, 1].imshow(binarim, cmap='gray')
axes[1, 0].imshow(1-dilate(1-binarim, selem=rectangle(8, 3)), cmap='gray')
axes[1, 1].imshow(1-dilate(1-binarim, selem=rectangle(20, 3)), cmap='gray')
for ax in axes:
    for a in ax:
        a.axis('off')
```

You can get some really funky behavior when you try exotic ones like stars:

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

axes[0, 0].imshow(image, cmap='gray')
binarim = image > 50
axes[0, 1].imshow(binarim, cmap='gray')
axes[1, 0].imshow(1-dilate(1-binarim, selem=star(3)), cmap='gray')
axes[1, 1].imshow(1-dilate(1-binarim, selem=star(7)), cmap='gray')
for ax in axes:
    for a in ax:
        a.axis('off')
```
